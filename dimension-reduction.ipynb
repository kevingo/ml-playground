## Curse of dimensionality

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
# load mnist data from datasets
digits = datasets.load_digits()
```

```python
# Create a regular PCA model 
pca = PCA(n_components=2)
tsne = TSNE(n_components=2)

# Fit and transform the data to the model
embedded_data_pca = pca.fit_transform(digits.data)
embedded_data_tsne = tsne.fit_transform(digits.data)

# Inspect the shape
print("Shape of embedded_data_pca:", embedded_data_pca.shape)
print("Shape of embedded_data_tsne:", embedded_data_tsne.shape)
print("---")

# Print out the data
print("PCA:")
print(embedded_data_pca)

print("TSNE:")
print(embedded_data_tsne)
```

```python
colors = ['black', 'blue', 'purple', 'yellow', 'white', 'red', 'lime', 'cyan', 'orange', 'gray']
for i in range(len(colors)):
    x = embedded_data_pca[:, 0][digits.target == i]
    y = embedded_data_pca[:, 1][digits.target == i]
    plt.scatter(x, y, c=colors[i])
plt.legend(digits.target_names, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title("PCA Scatter Plot")
plt.show()
```

```python
colors = ['black', 'blue', 'purple', 'yellow', 'white', 'red', 'lime', 'cyan', 'orange', 'gray']
for i in range(len(colors)):
    x = embedded_data_tsne[:, 0][digits.target == i]
    y = embedded_data_tsne[:, 1][digits.target == i]
    plt.scatter(x, y, c=colors[i])
plt.legend(digits.target_names, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title("PCA Scatter Plot")
plt.show()
```
